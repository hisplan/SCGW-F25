# Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load

```python
df = pd.read_csv("./data/diabetes.csv")
df
```

## Histogram

```python
fig, ax = plt.subplots(figsize=(6 ,6))

sns.histplot(df.Age, ax=ax)

plt.show()
```

## Box Plot

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.boxplot(x="Age", y="Glucose", data=df, ax=ax)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.boxplot(x="Age", y="Glucose", hue="Outcome", data=df[ df.Age < 50 ], ax=ax)

plt.show()
```

## Correlation

```python
df.corr()
```

## Heatmap

```python
fig, ax = plt.subplots(figsize=(6, 6))

cmap = sns.color_palette("YlOrBr", as_cmap=True)
sns.heatmap(df.corr(), cmap=cmap, ax=ax)

plt.show()
```
